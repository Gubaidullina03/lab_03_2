```python
# Задание task_03_02_01.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def invert_dictionary(d):

    inverted = {}
    for key, value in d.items():
        inverted[value] = key
    return inverted


product_codes = {'Laptop': 'A123', 'Mouse': 'B456', 'Keyboard': 'C789'}
print(invert_dictionary(product_codes))
```

```python
# Задание task_03_02_02.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def count_char_frequency(s):

    frequency = {}
    for char in s:
        if char in frequency:
            frequency[char] += 1
        else:
            frequency[char] = 1
    return frequency


print(count_char_frequency("hello world"))
```

```python
# Задание task_03_02_03.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def find_common_elements(list1, list2):

    set1 = set(list1)
    set2 = set(list2)
    common_elements = set1 & set2  # или set1.intersection(set2)
    return list(common_elements)


clients1 = ["Иванов", "Петров", "Сидоров", "Алексеев"]
clients2 = ["Сидоров", "Кузнецов", "Иванов", "Попов"]
print(find_common_elements(clients1, clients2))
```

```python
# Задание task_03_02_04.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def merge_dictionaries(dict1, dict2):

    result = dict1.copy()  # Создаем копию первого словаря
    result.update(dict2)   # Обновляем значения из второго словаря
    return result


prices1 = {'apple': 1.5, 'banana': 0.75}
prices2 = {'banana': 0.85, 'orange': 1.25}
print(merge_dictionaries(prices1, prices2))
```

```python
# Задание task_03_02_05.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def find_difference(list1, list2):

    set1 = set(list1)
    set2 = set(list2)
    difference = set1 - set2  # или set1.difference(set2)
    return list(difference)


skills_required = ["python", "sql", "git", "data_analysis"]
skills_candidate = ["python", "git", "excel"]
print(find_difference(skills_required, skills_candidate))
```

```python
# Задание task_03_02_06.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def filter_by_price(products, price_limit):

    result = []
    for product, price in products.items():
        if price > price_limit:
            result.append(product)
    return result


product_prices = {'Laptop': 1200, 'Mouse': 25, 'Keyboard': 75, 'Monitor': 300}
print(filter_by_price(product_prices, 100))
```

```python
# Задание task_03_02_07.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def group_by_department(employees):

    departments = {}
    for name, department in employees:
        if department not in departments:
            departments[department] = []
        departments[department].append(name)
    return departments


employee_list = [("Иванов", "IT"), ("Петров", "Sales"), ("Сидорова", "IT"), ("Кузнецов", "Marketing"), ("Алексеева", "Sales")]
print(group_by_department(employee_list))
```

```python
# Задание task_03_02_08.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def count_unique_words(text):

    # Приводим текст к нижнему регистру и разбиваем на слова
    words = text.lower().split()

    # Удаляем знаки препинания и создаем множество уникальных слов
    cleaned_words = []
    for word in words:
        # Убираем знаки препинания вокруг слова
        cleaned_word = word.strip('.,!?;:"()[]')
        if cleaned_word:  # Проверяем, что слово не пустое
            cleaned_words.append(cleaned_word)

    # Возвращаем количество уникальных слов
    return len(set(cleaned_words))


text = "Это пример текста для примера подсчета уникальных слов. Текста для примера."
print(count_unique_words(text))
```

```python
# Задание task_03_02_09.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def find_most_expensive(products):

    if not products:
        return None

    most_expensive_product = None
    highest_price = float('-inf')

    for product, price in products.items():
        if price > highest_price:
            highest_price = price
            most_expensive_product = product

    return most_expensive_product


product_prices = {'Laptop': 1200, 'Mouse': 25, 'Keyboard': 75, 'Monitor': 300}
print(find_most_expensive(product_prices))
```

```python
# Задание task_03_02_10.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def find_symmetric_difference(list1, list2):

    set1 = set(list1)
    set2 = set(list2)
    symmetric_diff = set1 ^ set2  # или set1.symmetric_difference(set2)
    return list(symmetric_diff)


store1_products = ["apple", "banana", "orange", "grape"]
store2_products = ["banana", "kiwi", "apple", "mango"]
print(find_symmetric_difference(store1_products, store2_products))
```

```python
# Задание task_03_02_11.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def get_nested_value(data_dict, key_path):

    current_level = data_dict

    for key in key_path:
        if isinstance(current_level, dict) and key in current_level:
            current_level = current_level[key]
        else:
            return None

    return current_level


user_profile = {
    'user_id': 101,
    'personal_info': {
        'name': 'Анна',
        'address': {
            'city': 'Москва',
            'street': 'Ленина'
        }
    }
}
print(get_nested_value(user_profile, ['personal_info', 'address', 'city']))
```

```python
# Задание task_03_02_12.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def update_inventory(inventory, new_items):
    # Создаем копию инвентаря, чтобы не изменять оригинал
    updated_inventory = inventory.copy()

    # Обходим все новые товары
    for item in new_items:
        if item in updated_inventory:
            # Увеличиваем количество существующего товара
            updated_inventory[item] += 1
        else:
            # Добавляем новый товар с количеством 1
            updated_inventory[item] = 1

    return updated_inventory


current_inventory = {'apples': 10, 'bananas': 15, 'oranges': 5}
new_shipment = ['apples', 'apples', 'oranges', 'grapes', 'bananas']
print(update_inventory(current_inventory, new_shipment))
```

```python
# Задание task_03_02_13.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def check_permissions(user_permissions, required_permissions):

    return required_permissions.issubset(user_permissions)


user_perms = {'read', 'write', 'comment'}
required_perms = {'read', 'write'}
print(check_permissions(user_perms, required_perms))
print(check_permissions(user_perms, {'read', 'delete'}))
```

```python
# Задание task_03_02_15.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def remove_below_threshold(data_dict, threshold):

    result = {}
    for key, value in data_dict.items():
        if value >= threshold:
            result[key] = value
    return result


stock_quantities = {'product_A': 150, 'product_B': 45, 'product_C': 200, 'product_D': 80}
print(remove_below_threshold(stock_quantities, 100))
```
